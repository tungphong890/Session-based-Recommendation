```python
# === Enhanced GRU4Rec + optional sampled softmax (for large vocab) ===
!pip install --quiet torch tqdm matplotlib

import os, json, shutil
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split

# --------------------------
# Config & paths
# --------------------------
DATA_DIR = "/kaggle/input/processeddatawebmining/WebMiningData"
OUT_DIR  = "/kaggle/working"
os.makedirs(OUT_DIR, exist_ok=True)

X_train = np.load(os.path.join(DATA_DIR, "X_train.npy"), allow_pickle=True)
y_train = np.load(os.path.join(DATA_DIR, "y_train.npy"), allow_pickle=True)
X_test  = np.load(os.path.join(DATA_DIR, "X_test.npy"),  allow_pickle=True)
y_test  = np.load(os.path.join(DATA_DIR, "y_test.npy"),  allow_pickle=True)

print("Train / test sizes:", len(X_train), len(X_test))

# build item → index mapping
all_items = set()
for seq in X_train:
    all_items.update(seq)
all_items.update(y_train.tolist())
item2idx = {item: idx+1 for idx, item in enumerate(sorted(all_items))}
VOCAB_SIZE = len(item2idx) + 1  # 0 for pad/unknown
print("Vocab size:", VOCAB_SIZE)

# hyperparams
EMBED_DIM = 256
HIDDEN_DIM = 256
DROPOUT = 0.2
BATCH_SIZE = 512
EPOCHS = 20
LR = 1e-3
WEIGHT_DECAY = 1e-5
GRAD_CLIP = 5.0
VALID_RATIO = 0.05
TOPK = 20
USE_SAMPLED = True       # whether to use sampled softmax
NEG_SAMPLE_COUNT = 100   # number of negative samples if sampled
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# --------------------------
# Dataset
# --------------------------
class SessionDataset(Dataset):
    def __init__(self, X, y, item2idx):
        self.X = X
        self.y = y
        self.item2idx = item2idx
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        seq = self.X[idx]
        mapped = [self.item2idx.get(i, 0) for i in seq]
        target = self.item2idx.get(self.y[idx], 0)
        return torch.tensor(mapped, dtype=torch.long), torch.tensor(target, dtype=torch.long)

def collate_fn(batch):
    seqs, targets = zip(*batch)
    lengths = torch.tensor([len(s) for s in seqs], dtype=torch.long)
    padded = nn.utils.rnn.pad_sequence(seqs, batch_first=True, padding_value=0)
    targets = torch.stack(targets)
    return padded, lengths, targets

dataset = SessionDataset(X_train, y_train, item2idx)
n_val = max(1, int(len(dataset)*VALID_RATIO))
train_ds, val_ds = random_split(dataset, [len(dataset)-n_val, n_val], generator=torch.Generator().manual_seed(42))
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_fn, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, collate_fn=collate_fn, num_workers=2, pin_memory=True)
test_ds = SessionDataset(X_test, y_test, item2idx)
test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, collate_fn=collate_fn, num_workers=2, pin_memory=True)

# --------------------------
# Model
# --------------------------
class GRURec(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, dropout):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.gru = nn.GRU(embed_dim, hidden_dim, batch_first=True)
        self.drop = nn.Dropout(dropout)
        self.out = nn.Linear(hidden_dim, vocab_size)
    def forward(self, seqs, lengths):
        e = self.emb(seqs)
        packed = nn.utils.rnn.pack_padded_sequence(e, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, h = self.gru(packed)
        h = h.squeeze(0)
        h = self.drop(h)
        logits = self.out(h)
        return logits

model = GRURec(VOCAB_SIZE, EMBED_DIM, HIDDEN_DIM, DROPOUT).to(device)

# --------------------------
# Loss / Optimizer
# --------------------------
optimizer = optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
criterion = nn.CrossEntropyLoss()  # or use sampled softmax / custom if USE_SAMPLED
# if you implement sampled softmax, you need custom loss + sampling logic (more complex)
# For now we keep full softmax; sampling is a more advanced step

# --------------------------
# Training + validation + test
# --------------------------
best_val = 0.0
history = {"train_loss": [], "val_loss": [], "val_recall": [], "val_mrr": []}

for epoch in range(1, EPOCHS+1):
    model.train()
    total_loss = 0.0
    for seqs, lengths, targets in tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS}"):
        seqs, lengths, targets = seqs.to(device), lengths.to(device), targets.to(device)
        optimizer.zero_grad()
        logits = model(seqs, lengths)
        loss = criterion(logits, targets)
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), GRAD_CLIP)
        optimizer.step()
        total_loss += loss.item()
    avg_loss = total_loss / len(train_loader)
    history["train_loss"].append(avg_loss)

    # validation
    model.eval()
    recall = 0.0; mrr = 0.0; total = 0
    with torch.no_grad():
        for seqs, lengths, targets in tqdm(val_loader, desc="Val", leave=False):
            seqs, lengths, targets = seqs.to(device), lengths.to(device), targets.to(device)
            logits = model(seqs, lengths)
            topk = torch.topk(logits, TOPK, dim=1)[1].cpu().tolist()
            t = targets.cpu().tolist()
            for preds, gt in zip(topk, t):
                total += 1
                if gt in preds:
                    recall += 1
                    rank = preds.index(gt) + 1
                    mrr += 1.0 / rank
    val_recall = recall / total
    val_mrr = mrr / total
    history["val_recall"].append(val_recall)
    history["val_mrr"].append(val_mrr)
    print(f"> Epoch {epoch} done. Train_loss={avg_loss:.4f}, Val Recall@{TOPK}={val_recall:.4f}, MRR={val_mrr:.4f}")
    if val_recall > best_val:
        best_val = val_recall
        torch.save(model.state_dict(), os.path.join(OUT_DIR, "best_gru.pth"))
        print("  Saved best model.")

# final test
model.load_state_dict(torch.load(os.path.join(OUT_DIR, "best_gru.pth")))
model.eval()
recall = 0.0; mrr = 0.0; total = 0
with torch.no_grad():
    for seqs, lengths, targets in tqdm(test_loader, desc="Test"):
        seqs, lengths, targets = seqs.to(device), lengths.to(device), targets.to(device)
        logits = model(seqs, lengths)
        topk = torch.topk(logits, TOPK, dim=1)[1].cpu().tolist()
        t = targets.cpu().tolist()
        for preds, gt in zip(topk, t):
            total += 1
            if gt in preds:
                recall += 1
                rank = preds.index(gt) + 1
                mrr += 1.0 / rank
test_recall, test_mrr = recall/total, mrr/total
print("FINAL TEST Recall@20:", test_recall, "MRR@20:", test_mrr)

# Save & plot
with open(os.path.join(OUT_DIR, "history_enhanced.json"), "w") as f:
    json.dump(history, f)
plt.figure(figsize=(8,4))
plt.plot(history["train_loss"], label="train_loss")
plt.title("Training Loss"); plt.legend(); plt.show()
plt.figure(figsize=(8,4))
plt.plot(history["val_recall"], label="val_recall@20")
plt.plot(history["val_mrr"], label="val_mrr@20")
plt.title("Validation Recall/MRR"); plt.legend(); plt.show()
```
